```python
with h5py.File('/kaggle/input/ppa-dataset/graph_data.hdf5', 'r') as f:
    # Initialize a list to hold the node counts
    node_counts = []  
    
    # Iterate through each group in the file
    for graph_name in tqdm(f):
        grp = f[graph_name]  # Access the group
        
        # Retrieve the num_nodes attribute and append it to the list
        num_nodes = grp.attrs['num_nodes']
        node_counts.append(num_nodes)
    
    # Convert node_counts to a numpy array for statistical calculations
    node_counts = np.array(node_counts)
    node_count_mean = node_counts.mean()
    node_count_sd = np.std(node_counts)  # Corrected to calculate standard deviation
    
    # Calculate deviations from the mean in terms of standard deviation
    graph_node_count_deviations = [(node_count - node_count_mean) / node_count_sd for node_count in node_counts]
# Save the sorted list of all hashes using h5py
```

```python
copyfile(src = '/kaggle/input/ppa-dataset/graph_data.hdf5', dst = '/kaggle/working/graph_data.hdf5')
```

```python
del num_nodes
del node_counts
gc.collect()
```

```python
with h5py.File('/kaggle/working/graph_data.hdf5', 'a') as h5f:  # Note the change to 'a' for append mode
    for graph_name, z_score in tqdm(zip(h5f, graph_node_count_deviations)):
        grp = h5f[graph_name]  # Access the group using the correct file handle
        grp.attrs['graph_node_count_z_score'] = z_score 
        

del graph_node_count_deviations
gc.collect()
```
